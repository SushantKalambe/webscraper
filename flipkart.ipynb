# Urls

```python
def links(url,tag,classes,domain):
    headers = {"User-Agent":"Mozilla/5.0"}
    result=requests.get(url,headers=headers).text
    soup=BeautifulSoup(result)
    all_links=soup.find_all(tag,classes)
    links=[]
    for i in all_links:
        links.append(i.attrs['href'])
    links=[domain + i for i in links]
    return links
```

```python
urls=links(url,'a','ge-49M',domain)
```

# inside page

```python
new_tag=['span','div','div','div','div']
new_classes=['B_NuCI','_30jeq3 _16Jk6d','_3I9_wc _2p6lqe','_3Ay6Sb _31Dcoz','_2418kt']
columns=['Mobile_mode','discount_price','orig-price','discount','description']
```

```python
def scrapper(url,links,tag,classes,columns):
    new_data=[]
    for i in links:
        headers = {"User-Agent":"Mozilla/5.0"}
        source=requests.get(i,headers=headers).text
        soupy=BeautifulSoup(source)
        for m,n in zip(new_tag,new_classes):
            soupy_2=soupy.find_all(m,n)
            if soupy_2:
                for j in soupy_2:
                    new_data.append((j.text).replace('\xa0', ''))
            else:
                new_data.append('None')
    nest=[]
    for k in range(0,len(new_data),len(new_classes)):
        nest.append((new_data[k:k+(len(new_classes))]))
    return (pd.DataFrame(nest,columns=columns))
```

```python
x=scrapper(url,links,new_tag,new_classes,columns)
```
